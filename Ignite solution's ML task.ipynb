# STEP 1

```python
mapper={'[28.0,-91.0]':'Region A',
 '[27.0,-81.0]':'Region B',
'[24.0,-101.0]':'Region C',
 '[41.0,-90.0]':'Region D',
 '[43.0,-75.0]':'Region E',
 '[31.0,-109.0]':'Region F',
 '[33.0,-105.0]':'Region G',
 '[33.0,-106.0]':'Region G',
 '[33.0,-85.0]':'Region H',
 '[44.0,-83.0]':'Region I',
 '[28.0,-75.0]':'Region J'}
```

```python
# tuples of all lats & longs of different regions

lst=[(28.0,-91.0),(27.0,-81.0),(24.0,-101.0),(41.0,-90.0),(43.0,-75.0),(31.0,-109.0),(33.0,-105.0),(33.0,-106.0),(33.0,-85.0),(44.0,-83.0),(28.0,-75.0)]

data=[]

def error_compute(x,y): #function to compute the minimum error & to map a location to it's nearest region
    
    root_vals=[]
    
    for i,j in lst:
        root_vals.append(((np.sqrt([(i-x)**2+(j-y)**2][0])),i,j)) #computing euclidean distance between the lat longs of location & the lat longs of various regions 
    soln=min(root_vals)[1:3]                                        # to find the minimum distance
    
    #print(soln)
    key='[{val1},{val2}]'.format(val1=soln[0],val2=soln[1]) # using string format for making the i,j values in string type so we can use them as a key to extract regions
    return(mapper[key])
```

```python
if __name__=="__main__":
    
    for i,j in zip(df['lat'],df['lon']):
        data.append(error_compute(i,j))
    region=pd.DataFrame(data=data,columns=['Region'])
    new_df1=pd.concat([df,region],axis=1) # A new df is created with shopping areas mapped to nearest regions 
```

### For grocery, the total monthly sales and the number of transactions per month for each region for each month in the period January 2018 to September 2020.

```python
grocery_2018=grocery_df[(grocery_df['year']==2018)&(grocery_df['month']>=1)&(grocery_df['month']<=12)]#df for whole year 2018

grocery_2019=grocery_df[(grocery_df['year']==2019)&(grocery_df['month']>=1)&(grocery_df['month']<=12)] #df for whole year 2019

grocery_2020=grocery_df[(grocery_df['year']==2020)&(grocery_df['month']>=1)&(grocery_df['month']<=9)]  #df for months jan-september 2020

step2_df=pd.concat([grocery_2018,grocery_2019,grocery_2020],axis=0)[['sale_amt','num_sales','month','year','Region']]

 
step2_df.groupby(by=['Region','month','year']).sum()[['sale_amt','num_sales']]
```

```python
grocery_2019=grocery_df[(grocery_df['year']==2019)&(grocery_df['month']>=1)&(grocery_df['month']<=12)] #df for whole year 2019

grocery_2020=grocery_df[(grocery_df['year']==2020)&(grocery_df['month']>=1)&(grocery_df['month']<=9)]  #df for months jan-september 2020

step3_df=pd.concat([grocery_2019,grocery_2020],axis=0)[['sale_amt','num_sales','month','year','Region']]

final_df1=step3_df.groupby(by=['Region','month','year']).sum()[['sale_amt','num_sales']]
```

### in the period Jan 2019 to Sep 2020, the percentage growth in monthly sales and the number of transactions for each month over the same month one year ago & also for n month of a year k to n-1 month of year k+1 .
#### n in range(1 to 12) & k in range(2019 to 2020)

#### Used pct_change method for claculating the percentage change via formulae: 
####     $$r_t = \frac{p_t - p_{t-1}}{p_{t-1}} = \frac{p_t}{p_{t-1}} - 1$$

```python
# Returns the percentage change for the current month sales & num of transactions w.r.t the same month for previous year

# As the df is multiindexed, the monthly sale amt & monthly num_transactions for consecutive year are one below other

final_df1.pct_change(periods=1)     
```

```python
final_df2=step2_df.groupby(by=['Region','month','year']).sum()[['sale_amt','num_sales']]

final_df3=final_df2.pct_change(periods=1)
```
